# Methodology

# <font size="4">Rules of the simulation

For this project we are not going to use any existing model nor theory, instead, we are going to design a simulation.

The simulation will follow some rules (or steps), which are going to be set in as much accordance as possible with our two main goals: 'Practicality' and 'Realism'. The 'Practicality' goal mainly refers to the fact that we want to create a simulation that addresses in a practical way the research questions formulated above.

We are not trying to solve any optimization problem, nor seeking to calculate an equilibrium in this simulation.

The rules (or steps) of our simulation will be the following:
1.	The simulation starts with a 'n' (5) number of individuals.
2.  Each initial individual is going to be assigned a level of ability to find food, which will be a random value from a minimum 'min_a' (1) to a maximum 'max_a' (3).
3.  Each initial individual is also going to be allocated to a community, which will be a random value from 1 to 'c' (5).
4.  Each initial individual will start the simulation with 0 savings of food.
5.  The simulation will run for 'p' (40) time periods.
6.	In each period, each individual finds a random amount of pieces of food, which will go from 0 to the individual's level of ability.
7.  There will be a completely exogenous and fixed amount of food 'max_f' (10) produced each period in the land where the simulation takes place.
8.	In case the sum of the amounts of food found by all the individuals in the simulation exceeds the total food produced, it means that the individuals will have to compete for some of the food they found, meaning some of them will not collect all the food they found. At the end, the sum of the food found by all the individuals will never exceed the total food produced.
9.	In case in the same community there is an individual with 0 food, and another individual with 2 or more food, the second individual will share 1 piece of food with the first individual. This happens repeatedly for every community until there are no more cases like the one mentionned right above.
10.	Every alive individual must consume 1 piece of food to survive and pass to the next period. Any individual that does not have at least 1 piece of food dies.
11.	Each individual has an exogenous probability 'l' (0.01) of leaving its community at the end of the period, joining any other alive community or creating a new one, each of the options with the same probability.
12.	At the end of each period, each individual has an exogenous probability 'r' of reproducing, creating a new individual with the same characteristics of its progenitor. However, both individuals will act independently.
13. Rules from 6 to 12 will take place each period until period 'p'.

# <font size="4">Main assumptions

1. All the parameters are constant and completely exogenous, arbitrarely chosen by us. This means output is exogenous and constant, meaning there is no technological or economic growth.
2. The simulation assumes 0 interaction with the outside world. Closed economy.
3. We have no information on any utility function or preferences that individuals try to maximize.
5. Individuals live forever unless they die from hunger.
6. Descendents are clones of their progenitors. There is no genetic mutation.
7. Food can be infinitely stored in amounts and periods of time, there is no stock depreciation.
8. Reproduction is carried out by individuals themselves, there is no need to join another individuals in order to reproduce.

# Python code

First, we import all the libraries we will need for the simulations and its further analyses:

```python
import random as rd
import numpy as np
import matplotlib.pyplot as mp
import collections as cl
```

Then, we start by setting the number of periods that each simulation will have:

```python
p=40
```

We continue by setting the number of individuals that will be alive at the start of each simulation:

```python
n=5
```

We set the minimum and the maximum levels of ability to find food that any individual can have in each simulation:

```python
min_a=1
max_a=3
```

The maximum amount of communities that each simulation will start with:

```python
c=5
```

The minimum and maximum amount of food (or savings) with which each individual can start with:

```python
min_s=0
max_s=0
```

The maximum amount of food that can be found in total by all the individuals each period:

```python
max_f=10
```

The chance that each alive individual has of leaving their community each period:

```python
l=0.01
```

And the chance that each alive individual has of reproducing each period:

```python
r=0.1
```

The next step is then to create an empty variable called 'ability', that will end up being a list in which the index of the list will represent each individual (sorted by age descending), and the value will indicate the ability that the individual has to find food:

```python
ability=None
```

We do also create another empty variable called 'community', that will end being a list of lists, in which the index of the outer list will represent each time period, the index of the inner lists will represent each individual (sorted by age descending), and the value will indicate in which community does the individual belong at the start of the indicated period:

```python
community=None
```

We similarly create an empty variable called 'savings', that will end being a list of lists, in which the index of the outer list will represent each time period, the index of the inner lists will represent each individual (sorted by age descending), and the value will indicate the amount of stored food that the individual has at the start of the indicated period:

```python
savings=None
```

And finally we create a last empty variable called 'food_found', that will end up being a list of lists, in which the index of the outer list will represent each time period, the index of the inner lists will represent each individual (sorted by age descending), and the value will indicate the amount of food that the individual has found during the indicated period:

```python
food_found=None
```

Now, we have to create one list for each previously mentioned empty variable, in which, at the end of each simulation, we will append the updated corresponding variable (e.g. we append the variable 'ability' to the list 'ability_data').

```python
ability_data=[]
community_data=[]
savings_data=[]
food_found_data=[]
```

The next step is to define as functions the different processes that will take place every simulation:

In the first place, we define as 'start' the function that sets the initial values needed to kick-off the simulation.

Note that it is necessary to call as 'global' the variables that we want to use (from outside the function) and modify (during the function). This is applied in all the functions when necessary.

We set the previously mentioned empty variables as empty lists, and start by setting their initial values: For 'ability', each individual will be independently assigned a random value from 'min_a' to 'max_a'; For 'community', each individual will start the first period being part of a random community from 1 to 'c'; And for 'savings', each individual will start the first period with a random value from 'min_s' to 'max_s'. We don't need to set an initial value for 'food_found':

```python
def start():
    global ability
    global community
    global savings
    global food_found
    ability=[]
    for i in range(n):
        ability.append(rd.randint(min_a,max_a))
    community=[[]]
    for i in range(n):
        community[0].append(rd.randint(1,c))
    savings=[[]]
    for i in range(n):
        savings[0].append(rd.randint(min_s,max_s))
    food_found=[]
```

In the second place, and starting to define the processes that will take place every simulation, every period, we define 'find_food' as a function.

Alive individuals find food depending on their ability level. They find a random amount of food that will go from 0 to their 'ability'. Not alive individuals will directly find 0 food.

Then, if the sum of food found by all individuals in the period happened to exceed the maximum amount of food 'max_f', the difference between the sum and 'max_f' would be randomly taken away, unit by unit, from all individuals whose 'food_found' in period 1 is higher than 0:

```python
def find_food(t):
    global ability
    global community
    global food_found
    food_found.append([])
    for i in range(len(ability)):
        if community[t][i]!=0:
            food_found[t].append(rd.randint(0,ability[i]))
        else:
            food_found[t].append(0)
    while max_f<sum(food_found[t]):
        chosen=rd.randint(0,len(ability)-1)
        if 0<food_found[t][chosen]:
            food_found[t][chosen]-=1
```

In the third place, we define 'store_and_share' as a new function intended to be called right after the 'find_food' function.

In this function we sum the savings that each individual started the period with, plus the amount of food found by the individual during the same period; and we store the resulting values as the savings with which the individual will start the next period.

Then, alive individuals with 0 pieces of food (that would die at the end of the period because they would be unable to eat), are given 1 piece of food by someone from the same community, given that this 'someone' has at least 2 pieces of food. Younger individuals would be prioritized in case of doubt (we accomplish thi by revearsing the second 'for' loop):

```python
def store_and_share(t):
    global ability
    global community
    global savings
    global food_found
    savings.append([])
    for i in range(len(ability)):
        savings[t+1].append(savings[t][i]+food_found[t][i])
    for i in reversed(range(len(ability))):
        if community[t][i]!=0 and savings[t+1][i]<1:
            for j in range(len(ability)):
                if community[t][j]==community[t][i] and 1<savings[t+1][j]:
                    savings[t+1][j]-=1
                    savings[t+1][i]+=1
                    break
```

In the fourth place, we define 'survive': Almost at the end of the period, individuals with at least 1 piece of food would consume 1 piece to survive, and individuals that don't have any food die:

```python
def survive(t):
    global ability
    global community
    global savings
    community.append([])
    for i in range(len(ability)):
        if 0<savings[t+1][i]:
            savings[t+1][i]-=1
            community[t+1].append(community[t][i])
        else:
            community[t+1].append(0)
```

In the fifth place, we define 'change_community': For every alive individual, there is a chance 'l' that she/he leaves their community, and if they do so, they join any other existing community or create a new one, all these options with equal probability:

```python
def change_community(t):
    global ability
    global community
    for i in range(len(ability)):
        if community[t+1][i]!=0 and rd.uniform(0,1)<l:
            while community[t+1][i]==community[t][i]:
                community[t+1][i]=rd.randint(1,max(community[t])+1)
```

And finally, we define 'reproduce' as a function: At the end of the period, each alive individual has a chance 'r' of reproducing, giving birth at the start of the next period to a new individual. We set the variables 'community', 'savings' and 'food_found' of the newborn individuals to 0 for all the previous periods previous to their birth; We set their variables 'ability' and 'community' (at the time of birth) as equal to its progenitors'; And we set 'savings' (at the time of birth) equal to 0:

```python
def reproduce(t):
    global ability
    global community
    global savings
    global food_found
    for i in range(len(ability)):
        if community[t+1][i]!=0 and rd.uniform(0,1)<r:
            for u in range(t+1):
                community[u].append(0)
                savings[u].append(0)
                food_found[u].append(0)
            ability.append(ability[i])
            community[t+1].append(community[t+1][i])
            savings[t+1].append(0)
```

Now we define the function that will run an entire simulation: from setting the bases of the simulation calling the 'start' function, to looping through all the time periods, and for each period calling each of the other previoulsy defined functions:

```python
def simulate():
    start()
    for t in range(p):
        find_food(t)
        store_and_share(t)
        survive(t)
        change_community(t)
        reproduce(t)
```

And the last step before analyzing the simulation-generated data, is to define a function that runs loops as many times as possible (to have a big enough sample where the law of large numbers applies), and in each loop, the previous function 'simulate' is called, and appends all the final generated data from the simulation to the lists that gather the entire dataset from all the simultions ran:

```python
def run_and_store(ability_data,community_data,savings_data,food_found_data):
    for x in range(100000):
        simulate()
        ability_data.append(ability.copy())
        community_data.append(community.copy())
        savings_data.append(savings.copy())
        food_found_data.append(food_found.copy())
```

We now have defined the necessary functions in order to run 100000 simulations and save all the generated data.

This enables us now to analyze the entire dataset, and so, to try to answer the research questions through data.

To answer the first research question, we define a function called 'com_con_time', in which we want to find the evolution through time of the average share of individuals that are part of the biggest community; i.e. we want to find the evolution of the average community concentration through time.

We run a 'for' loop inside another 'for' loop, for every combination of simulations and time periods, in order to filter all the 'community_data', to retrieve only the data on the alive individuals. We do then count the people that there is in each community, pick the amount of people that there is in the largest community, and divide it between the total number of people alive: this results in the share of individuals that are part of the biggest community, which is a proxy of the community concentration. If there are no alive individuals in a simulation, its community concentration cannot be really calculated, so that piece of data is dropped. Once we have this newly created data, we calculate the average of all the simulations for each period, resulting in a list called 'av_com_concentration'.

To make the results clear, we graph the resulting 'av_com_concentration' as a line graph so that we can visualize which is the evolution; and on top of that, we calculate the correlation between the average community concentration and time, so that we can know whether the correlation is positive or negative:

```python
def com_con_time():
    com_concentration=[]
    av_com_concentration=[]
    for t in range(p):
        com_concentration.append([])
        for x in range(len(community_data)):
            com_alive=list(filter(lambda x:x!=0,community_data[x][t]))
            count_com=cl.Counter(com_alive)
            try:
                com_concentration[t].append(max(count_com.values())/len(com_alive))
            except:
                continue
        av_com_concentration.append(np.average(com_concentration[t]))
    print('The correlation coefficient between community concentration and time: '+str(np.corrcoef(av_com_concentration,range(p))[0][1]))
    mp.title('Community concentration through time')
    mp.xlabel('Time periods')
    mp.ylabel('Average community concentration')
    mp.plot(av_com_concentration)
    mp.show()
```

To answer the second research question, we define a function called 'com_endowment', in which we want to know how the initial endowment on the community concentration of a simulation affects the population size in the short and in the long run.

We run a 'for' loop that runs through every simulation, and for each simulation it calculates: the initial community concentration; the population at period 'p'/10 (short run); and the popualtion at period 'p' (long run). We then plot two scatter plots in which each simulation is a dot, that indicates through the x-axis the initial community concentration, and through the y-axis of the first graph the population size at period 'p'/10 and through the y-axis of the second graph the population size at period 'p'. We finally represent in each scatter plot its corresponding correlation or 'best fit' line:

```python
def com_endowment():
    com_con_endowment=[]
    pop_p4=[]
    pop_p40=[]
    for x in range(len(community_data)):
        com_alive=list(filter(lambda x:x!=0,community_data[x][0]))
        count_com=cl.Counter(com_alive)
        com_con_endowment.append(max(count_com.values())/n)
        pop_p4.append(len(list(filter(lambda x:x!=0,community_data[x][int(p/10)]))))
        pop_p40.append(len(list(filter(lambda x:x!=0,community_data[x][p]))))
    fig=mp.figure()
    ax1=fig.add_subplot(221)
    ax1.set_title('(At p='+str(int(p/10))+')')
    ax1.set_xlabel('Initial community concentration')
    ax1.set_ylabel('Population')
    ax1.scatter(com_con_endowment,pop_p4,s=0.1)
    ax1.plot(np.unique(com_con_endowment),np.poly1d(np.polyfit(com_con_endowment,pop_p4,1))(np.unique(com_con_endowment)),color='red')
    ax2=fig.add_subplot(222)
    ax2.set_title('(At p='+str(p)+')')
    ax2.set_xlabel('Initial community concentration')
    ax2.scatter(com_con_endowment,pop_p40,s=0.1)
    ax2.plot(np.unique(com_con_endowment),np.poly1d(np.polyfit(com_con_endowment,pop_p40,1))(np.unique(com_con_endowment)),color='red')
    mp.show()
   
```

Last but not least, to answer the third research question, we define a simple function called 'steady_state', in which we want to know how does the total population size evolve through time on average.

We run a 'for' loop inside another 'for' loop, for every combination of simulations and time periods, in order to filter all the 'community_data', to retrieve only the data on the alive individuals. We store all the data, and we then calculate the average alive population from all the simulations, for every period. We then plot a simple line graph so that we can visually analyze if there is a population steady state to which simulations tend to arrive on average:

```python
def steady_state():
    av_pop=[]
    pop=[]
    for t in range(p):
        pop.append([])
        for x in range(len(community_data)):
            pop[t].append(len(list(filter(lambda x:x!=0,community_data[x][t]))))
        av_pop.append(np.average(pop[t]))
    mp.title('Population through time')
    mp.xlabel('Time periods')
    mp.ylabel('Average population')
    mp.plot(av_pop,color='black')
    mp.show()
```

# Discussion and conclusion

As seen in the last graph 'Figure 10', we can conclude that our first hypothesis heavily relies on the arbitrary determination of the initial parameters. Specifically, we have seen that it heavily relies on the value of 'l', which is a completely exogenous rate, which seems to be in reality highly endogenous (as it relies on emotions, ideology, amount of people in the community, personal benefit, etc). It would be very interesting to define utility functions for the willingness to leave and/or join communities. This way, individuals would choose to leave their community, to act on their own, or to join another one with higher chances of success. It was difficult for us to arbitrarely determine a rate for 'l', so we don't have a splendid case to defent the chosen value of 0.01. For further research, if the exogeneity of this parameter is still necessary, we could try to establish a believeable rate taking into account past theoretical or empirical research on the topic.

About the surprising result from the second hypothesis shown with the negative correlation in 'Figure 3', after checking the same graph (and some raw data) but changing the levels of ability during the sensitivity analysis, we realized that a possible explanation of what happenned was that when there are individuals of different abilities, when they are not united in a large community, rapidly the ones with low abilities die out quickly and only the ones with high abilities survive, rapidly enforcing natural selection and creating a society of only individuals with high abilities. On the other hand, when there are individuals of different abilities but they are all united under a large community, if there are enough individuals with high abilities, the individuals with low abilities survive as well, eliminating natural selection, and creating a society of individual of both low and high abilities, being less efficient that the one previously mentioned, leading to less population in the long run.

This is an interesting result that would need more data and more sensitivity analyses to be confirmed, but still feels to be true when observing how our societies seem not to be evolving through natural selection anymore, but thorugh other mechanisms.

Finally, and taking into account the results and sensitivity analyses tackling the 3rd research question, we can conclude that one of the main weaknesses of our approach is that the population steady state we have found is absolutely dependent on total output, which, in our case, is exogenoulsy given and constant through time. It would be very interesting to run the same simulation but defining a production function with dynamic endogenous inputs instead of an exogenous fixed amount. Running such simulation would enable to analyze better the effects of communities and individualism in a society.
