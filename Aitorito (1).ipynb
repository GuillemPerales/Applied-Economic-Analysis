In the long run, we expect that population size is going to reach a kind of steady state (maximum population size) independently from the initial endowment of individuals concentration in a large community. This is the hypothesis that we expect to answer our third research question.

# Methodology 

# <font size="4">Rules of the game

In our simulation we are not going to use any existing model nor theory. Instead, we are going to design a dynamic 'game' which's rules are going to be set in order to address the research questions formulated above. Hence, we are not trying to solve any optimization problem nor seeking to calculate an equilibrium. However, we have left as a research question whether the simulations lead to steady state (an equilibrium) or to different results depending on the initial endowment. The rules of the simulation game are the following:
1.	The simulation starts with 'n' number of individuals. Each individual 'i' is going to be randomly allocated to a community 'c’ and is going to have a level of ability to find food 'a'. The range of 'c' communities goes from 0 to infinite and the ability to find food 'a' is going to determine how many pieces of food can an individual find in each period at most with a minimum of ability 'min_a' value of 1 and a maximum 'max_a' of 3.
2.	In each period 'p', each individual finds 'f' pieces of food (which can go from 0 to 'a'). However, the available amount of food in each period is going to be ‘max_f’, which is going to be set arbitrarily by us.
3.	We are going to simulate ‘p’ periods of time.
4.	Each individual 'i' starts with a random amount of savings between 'min_s' and 'max_s'. The variable savings stands for the number of pieces of food 'f' that an individual was able to save for the next period. Therefore, savings in each period are going to be equal to the savings accumulated for the previous period (p-1), plus 'f' pieces of food found in the previous period (p-1), minus 1 piece of food which was needed to be consumed in order to survive to the next period (in p-1). Consequently, in each period, each individual has a number of pieces of food equal to the savings achieved for the period plus the found pieces of food in the current period.
5.	In the case that in a community there is an individual (or some individuals) with more than 1 piece of food, the exceed can be redistributed to individuals without any pieces of food. Therefore, the number of pieces of food in each period is going to be equal to the savings for the current period, plus, the pieces of food found in the current period, plus, one piece of food coming from another member of the community (in case that anyone has savings+found food>1, and another one has savings+found food=0)
6.	Individuals with 0 pieces of food die in a period 'p' are not going to be able to consume the necessary unit of food to survive to the next period and, consequently, is going to die.
7.	Each individual has a probability 'l' of leaving its community in each period of time.
8.	Each individual has a probability 'r' of reproducing, creating a new individual with the same characteristics of its progenitor. However, once the individual is born, it's going to act independently from its progenitor.

 # <font size="4">Main assumptions

1. Output is exogenously given and constant over time. Consequently, output is not determined by any production factor nor technological improvements, and there is not economic growth.
2. The game assumes closed frontiers, indivuduals in the society can't obtain food from other societies.
3. Individuals have no information on other individuals.
4. Leaving or joining a community is exogenously determined by a probability 'l' and 'j' respectively. Therefore, individuals can't undertake any choice, avoiding thus the possiboility to reach a dominant strategy or any equilibrium choices.
5. Individuals live forever unless they end up in having 0 pieces of food in any period.
6. Descendents are clones of their progenitors.
7. Food can be infinitely stored in amount and periods of time, there is no depreciation.
8. Reproduction is carried out by individuals themselves, there is no need to join another individual in order to reproduce.


# Python code

```python
import random as rd
import numpy as np
import matplotlib.pyplot as mp
import collections as cl
```

```python
p=40
n=5
min_a=1
max_a=3
c=5
min_s=0
max_s=0
max_f=10
l=0.01
r=0.1
ability=None
community=None
savings=None
food_found=None
ability_data=[]
community_data=[]
savings_data=[]
food_found_data=[]
```

```python
def start():
    global ability
    global community
    global savings
    global food_found
    ability=[]
    for i in range(n):
        ability.append(rd.randint(min_a,max_a))
    community=[[]]
    for i in range(n):
        community[0].append(rd.randint(1,c))
    savings=[[]]
    for i in range(n):
        savings[0].append(rd.randint(min_s,max_s))
    food_found=[]
```

```python
def find_food(t):
    global ability
    global community
    global food_found
    food_found.append([])
    for i in range(len(ability)):
        if community[t][i]!=0:
            food_found[t].append(rd.randint(0,ability[i]))
        else:
            food_found[t].append(0)
    while max_f<sum(food_found[t]):
        chosen=rd.randint(0,len(ability)-1)
        if 0<food_found[t][chosen]:
            food_found[t][chosen]-=1
```

```python
def store_and_share(t):
    global ability
    global community
    global savings
    global food_found
    savings.append([])
    for i in range(len(ability)):
        savings[t+1].append(savings[t][i]+food_found[t][i])
    for i in reversed(range(len(ability))):
        if community[t][i]!=0 and savings[t+1][i]<1:
            for j in range(len(ability)):
                if community[t][j]==community[t][i] and 1<savings[t+1][j]:
                    savings[t+1][j]-=1
                    savings[t+1][i]+=1
                    break
```

```python
def survive(t):
    global ability
    global community
    global savings
    community.append([])
    for i in range(len(ability)):
        if 0<savings[t+1][i]:
            savings[t+1][i]-=1
            community[t+1].append(community[t][i])
        else:
            community[t+1].append(0)
```

```python
def change_community(t):
    global ability
    global community
    for i in range(len(ability)):
        if community[t+1][i]!=0 and rd.uniform(0,1)<l:
            while community[t+1][i]==community[t][i]:
                community[t+1][i]=rd.randint(1,max(community[t])+1)
```

```python
def reproduce(t):
    global ability
    global community
    global savings
    global food_found
    for i in range(len(ability)):
        if community[t+1][i]!=0 and rd.uniform(0,1)<r:
            for u in range(t+1):
                community[u].append(0)
                savings[u].append(0)
                food_found[u].append(0)
            ability.append(ability[i])
            community[t+1].append(community[t+1][i])
            savings[t+1].append(0)
```

```python
def simulate():
    start()
    for t in range(p):
        find_food(t)
        store_and_share(t)
        survive(t)
        change_community(t)
        reproduce(t)
```

```python
def run_and_store(ability_data,community_data,savings_data,food_found_data):
    for x in range(10000):
        simulate()
        ability_data.append(ability.copy())
        community_data.append(community.copy())
        savings_data.append(savings.copy())
        food_found_data.append(food_found.copy())
```

```python
def com_con_time():
    com_concentration=[]
    av_com_concentration=[]
    for t in range(p):
        com_concentration.append([])
        for x in range(len(community_data)):
            com_alive=list(filter(lambda x:x!=0,community_data[x][t]))
            count_com=cl.Counter(com_alive)
            try:
                com_concentration[t].append(max(count_com.values())/len(com_alive))
            except:
                continue
        av_com_concentration.append(np.average(com_concentration[t]))
    mp.title('Community concentration through time')
    mp.xlabel('Time periods')
    mp.ylabel('Average community concentration')
    mp.plot(av_com_concentration)
    mp.show()
    print('Correlation coefficient between community concentration and time: '+str(np.corrcoef(av_com_concentration,range(p))[0][1]))
```

```python
def com_endowment():
    com_con_endowment=[]
    pop_p4=[]
    pop_p40=[]
    for x in range(len(community_data)):
        com_alive=list(filter(lambda x:x!=0,community_data[x][0]))
        count_com=cl.Counter(com_alive)
        com_con_endowment.append(max(count_com.values())/n)
        pop_p4.append(len(list(filter(lambda x:x!=0,community_data[x][int(p/10)]))))
        pop_p40.append(len(list(filter(lambda x:x!=0,community_data[x][p]))))
    fig=mp.figure()
    ax1=fig.add_subplot(221)
    ax1.set_title('(At p='+str(p/10)+')')
    ax1.set_xlabel('Initial community concentration')
    ax1.set_ylabel('Population')
    ax1.scatter(com_con_endowment,pop_p4,s=0.1)
    ax1.plot(np.unique(com_con_endowment),np.poly1d(np.polyfit(com_con_endowment,pop_p4,1))(np.unique(com_con_endowment)),color='red')
    ax2=fig.add_subplot(222)
    ax2.set_title('(At p='+str(p)+')')
    ax2.set_xlabel('Initial community concentration')
    ax2.scatter(com_con_endowment,pop_p40,s=0.1)
    ax2.plot(np.unique(com_con_endowment),np.poly1d(np.polyfit(com_con_endowment,pop_p40,1))(np.unique(com_con_endowment)),color='red')
    mp.show()
   
```

```python
def steady_state():
    av_pop=[]
    pop=[]
    for t in range(p):
        pop.append([])
        for x in range(len(community_data)):
            pop[t].append(len(list(filter(lambda x:x!=0,community_data[x][t]))))
        av_pop.append(np.average(pop[t]))
    mp.title('Population through time')
    mp.xlabel('Time periods')
    mp.ylabel('Average population')
    mp.plot(av_pop,color='black')
    mp.show()
```
